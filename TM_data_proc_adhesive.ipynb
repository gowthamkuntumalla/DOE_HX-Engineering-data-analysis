```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import os
from scipy.interpolate import make_interp_spline, BSpline
import scipy.signal as sci_sg
def list_files(dir):
    r = []
    direc = []
    for root, dirs, files in os.walk(dir):
        for name in files:
            r.append(os.path.join(root, name))
            if 'Cd1' in name:
                direc.append(name)
    return r, direc
#np.size(r)
```

```python
#Import data
root_dirc = r'D:\Tensile Test\MengFeb7'
r, direc = list_files(root_dirc)
data = []
calib_data = [ ]
titles= []
n = 0;
i=0
for i in range(0,np.size(r)):
    if 'Cd1' in r[i]:
        dataf = pd.read_csv(r[i],delimiter='\t', header = 0, index_col = None)
        data.append(np.asarray(dataf.reset_index().values)[1:])
        titles.append(r[i])
        n+=1 
        #print(n)
        #data_ar = np.array([np.vstack(np.array(xi)).astype(np.float) for xi in data])
```

```python
# Rearranging test data
data_ar =[]
for xi in data:
    data_ar.append(np.vstack(np.array(xi)).astype(np.float))
data_ar = np.array(data_ar)    
#print(data_ar)
```

```python
# Calibration data
calib_dataf = pd.read_csv(root_dirc+"\\clampedd1.txt",delimiter='\t', header = 0, index_col = None)
calib_data_ar = np.asarray(calib_dataf.reset_index().values[1:].astype(np.float))
Y_fit = sci_sg.savgol_filter(calib_data_ar[:,1] - min(calib_data_ar[:,1]) , 25, 5 )   # Savitzky Golay filter for improving SNR
y_calib = 1000*(np.mean(Y_fit))
print(str(y_calib) + 'N')
```

```python
# Plotting Peel and Shear Tests
Peel_str = [] # nested list [ P test, T, avg value in N/cm, max, min ]
Shear_str = [] # nested list [ S test, T, avg value in MPa, max, min]
TM_type = ["Acr_P_1milVS5mil","Acr_S_1milVS5mil","Epox_P_1milVS1mil","Epox_P_6milVS1mil", "Epox_S_1milVS1mil","Epox_S_6milVS1mil",
          "Sili_P_1milVS1mil","Sili_P_6milVS5mil", "Sili_S_1milVS1mil", "Sili_S_6milVS5mil"] # Thermomechanical test type
Temps = ['22C','100C','150C','200C'] # Temperature
clrs = ['b','r','g']

for test in TM_type:
    inds= [i for i in range(n) if test in titles[i]] # data indices for this test
    for T in Temps: # for each temperaure
        inds_T = [i for i in inds if T in titles[i]]
        peel_val_T = []
        shear_val_T = []
        if len(inds_T) ==0:
            continue
        plt.figure(figsize=(16,10))
        for j in inds_T: # plot all repeat runs in one plot
            X = data_ar[j][:,0]-min(data_ar[j][:,0])
            Y = 1000*(data_ar[j][:,1]-min(data_ar[j][:,1])) 
            Y_fit = sci_sg.savgol_filter( Y, 25, 5 )   # Savitzky Golay filter for improving SNR
            plt.plot( X ,Y_fit, clrs[inds_T.index(j)]) 
            
            if "_P" in test:
                print("In "+ test+ " @ " + T + " Peel Strength = "+ str(round(max(Y_fit)/1.5,2)) + " N/cm" ) 
                peel_val_T.append(round(max(Y_fit)/1.5,2))
                # required > 3.9 N/cm as per proposal
            if "_S" in test:
                print("In "+ test+ " @ " + T + " Max Shear Strength = "+ str(round(max(Y_fit)/(1.5*2.54*100),3)) + " MPa" )    
                shear_val_T.append(round(max(Y_fit)/(1.5*2.54*100),2))
        plt.xlabel("Pos (mm)")
        plt.ylabel("Load (N)"  )
        plt.title(test+ " @ " + T)
        plt.show()   
        
        # Avg strength data for this test @ T
    
        if len(peel_val_T) !=0:
            Peel_str.append([test,T,round(sum(peel_val_T)/len(peel_val_T),2), max(peel_val_T),min(peel_val_T)])
        if len(shear_val_T) !=0:
            Shear_str.append([test,T,round(sum(shear_val_T)/len(shear_val_T),2),max(shear_val_T),min(shear_val_T)])    
           
```

```python
# Rename all the title (Prof. Sinha told me to change the name)

cvtnm = {'Acr_P_1milVS5mil':'Acrylic', 'Epox_P_1milVS1mil':'Epoxy 1', 'Epox_P_6milVS1mil': 'Epoxy 2',
        'Sili_P_1milVS1mil': 'Silicone 1', 'Sili_P_6milVS5mil': 'Silicone 2', 'Acr_S_1milVS5mil': 'Acrylic',
        'Epox_S_1milVS1mil': 'Epoxy 1', 'Epox_S_6milVS1mil': 'Epoxy 2', 'Sili_S_1milVS1mil': 'Silicone 1',
        'Sili_S_6milVS5mil': 'Silicone 2'}
```

```python
clors = ['k', 'g', 'b', 'c']
linsty = ['--o', '-.o', ':o', '-o']
# peel strength avg values
j = 0
# fig = plt.figure(figsize= (14,6))
for T in Temps:
    inds_T = [i for i in  range(len(Peel_str)) if Peel_str[i][1]==T]

    X = [cvtnm[Peel_str[i][0]] for i in inds_T]
    Y = [Peel_str[i][2] for i in inds_T] 
    c = [round((Peel_str[i][3] - Peel_str[i][4])/2,2) for i in inds_T]
    if T == '150C':
        X = X[1:]
        Y = Y[1:]
        c = c[1:]
#     plt.plot(X,Y, linsty[j], color = clors[j], label = ('Temp = ' + T[:-1] +'$^\circ C$'))
    plt.errorbar(X,Y, yerr = c, linestyle = linsty[j][:-1], fmt ='-o'+clors[j],ecolor= clors[j], 
             mfc = clors[j],capsize= 5, label = ('Temp = ' + T[:-1] +'$^\circ C$'))
    j = j + 1
plt.axhline(y = 3.9, color='r', linestyle='-',label = 'Proposal')
# plt.axhline(y = y_calib/1.5, color='y', linestyle='--',label = 'calibration er')

# plt.xticks(fontsize = 15)
# plt.yticks(fontsize = 15)
plt.xlabel("Test Type")
plt.ylabel("Peel Strength (N/cm)")
plt.title(" Peel Strength of Adhesion ")  
plt.legend(fontsize = 8, loc = 1)
#saving figure
fname = root_dirc +"\\Peel_str"+ ".jpg"
plt.savefig(fname, format = 'jpg', dpi = 500 )
plt.show()  

    
# Shear T in Temps:
j = 0
# fig = plt.figure(figsize= (14,6))
for T in Temps:
    inds_T = [i for i in  range(len(Shear_str)) if Shear_str[i][1]==T]
    
    X = [cvtnm[Shear_str[i][0]] for i in inds_T]
    Y = [Shear_str[i][2] for i in inds_T] 
    c = [round((Shear_str[i][3] - Shear_str[i][4])/2,2) for i in inds_T]
#     plt.plot(X,Y, linsty[j], color = clors[j], label = ('Temp = ' + T[:-1] + '$^\circ C$'))
    plt.errorbar(X,Y, yerr = c, linestyle = linsty[j][:-1], fmt ='-o'+clors[j],ecolor= clors[j], 
                 mfc = clors[j], capsize= 5, label = ('Temp = ' + T[:-1] +'$^\circ C$'))
    
    j = j + 1
    
# plt.axhline(y=1, color='r', linestyle='-', label = '1Mpa')  
# plt.axhline(y = y_calib/(1.5*2.54*100), color='y', linestyle='--',label = 'calibration er')
# plt.xticks(fontsize = 15)
# plt.yticks(fontsize = 15)
plt.ylim([0, 0.7])
plt.xlabel("Test Type")
plt.ylabel("Max Shear Strength (MPa)")
plt.title(" Shear Strength of Adhesion")  
plt.legend(fontsize = 8, loc = 1)
#saving figure
fname = root_dirc +"\\Shear_str"+ ".jpg"
plt.savefig(fname, format = 'jpg', dpi = 500 )
plt.show()   

```
